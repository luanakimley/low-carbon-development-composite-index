# Normalisation

```python
import pandas as pd
```

```python
lcdi_df = pd.read_csv('lcdi/lcdi_final.csv')
lcdi_df
```

Implement min-max normalisation with the formula:
$Xnorm = \frac{X-Xmin}{Xmax-Xmin}$


```python
def min_max_normalise(df, column):
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())
    return df 
```

```python

lcdi_df = min_max_normalise(lcdi_df, 'forest_area')
lcdi_df = min_max_normalise(lcdi_df, 'freshwater')
lcdi_df = min_max_normalise(lcdi_df, 'years_since_unfccc')
lcdi_df = min_max_normalise(lcdi_df, 'renewable_energy_consumption')
lcdi_df = min_max_normalise(lcdi_df, 'total_greenhouse_gas_emission')
lcdi_df
```

```python
lcdi_df.to_csv("lcdi/lcdi_normalised.csv", index=False)
```
