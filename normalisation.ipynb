# Normalisation

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
lcdci_df = pd.read_csv('lcdci_final.csv')
lcdci_df
```

```python
current_year = 2024
lcdci_df["unfccc"] = current_year - lcdci_df['unfccc']
lcdci_df
```

```python
def min_max_normalise(df, column):
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())
    return df 
```

```python

lcdci_df = min_max_normalise(lcdci_df, 'forest_area')
lcdci_df = min_max_normalise(lcdci_df, 'freshwater')
lcdci_df = min_max_normalise(lcdci_df, 'unfccc')
lcdci_df = min_max_normalise(lcdci_df, 'renewable_energy_consumption')
lcdci_df = min_max_normalise(lcdci_df, 'total_greenhouse_gas_emission')
lcdci_df
```

```python
lcdci_df.to_csv("lcdci_normalised.csv", index=False)
```
